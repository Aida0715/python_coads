```python
#hstファイルを読み各エネルギーの保存を確認するコード

import numpy as np
import matplotlib.pyplot as plt

# ============================================
# hst ファイル読み込み
# ============================================

file = "/home/aian/athenapp/results/〇〇/〇〇.hst"

data = np.loadtxt(file)

time = data[:,0]
dt   = data[:,1]
mass = data[:,2]

momx = data[:,3]
momy = data[:,4]
momz = data[:,5]

KE1  = data[:,6]
KE2  = data[:,7]
KE3  = data[:,8]

Etot = data[:,9]
Egrav = data[:,10]

# ============================================
# エネルギー計算
# ============================================

KE = KE1 + KE2 + KE3

# 真の全エネルギー
E_total = Etot + Egrav

# 初期値
E0 = E_total[0]

# 相対誤差
E_err = (E_total - E0) / abs(E0)　　#初期エネルギーからのズレ（数値計算によるエネルギー非保存の指標）

# ============================================
# 基本情報
# ============================================

print("Initial mass:", mass[0])
print("Final mass:", mass[-1])

print("\nInitial total energy:", E_total[0])
print("Final total energy:", E_total[-1])

print("\nRelative energy error:", E_err[-1])

# ============================================
# プロット
# ============================================

plt.figure(figsize=(10,6))

plt.plot(time, KE, label="Kinetic Energy")
plt.plot(time, Egrav, label="Gravitational Energy")
plt.plot(time, Etot, label="Hydro Energy")
plt.plot(time, E_total, label="Total Energy")

plt.xlabel("Time")
plt.ylabel("Energy")

plt.legend()
plt.grid()

plt.tight_layout()
plt.show()

# ============================================
# エネルギー誤差
# ============================================

plt.figure(figsize=(10,5))

plt.plot(time, E_err)

plt.xlabel("Time")
plt.ylabel("Relative Energy Error")

plt.grid()
plt.tight_layout()
plt.show()
```
